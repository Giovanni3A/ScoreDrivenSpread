# Converter Dollars per Barrel para Dollars per Gallon

How many gallons are in a barrel?
Barrel (unit) - Wikipedia
42 US gallons

Definitions. In the oil industry, one barrel (unit symbol bbl) is a unit of volume used for measuring oil defined as approximately 42 US gallons or 160 litres or 35 imperial gallons.

```python
# Converter Dollars per Barrel para Dollars per Gallon
df_dados['Cushing, OK WTI Spot Price FOB (Dollars per Gallon)'] = df_dados['Cushing, OK WTI Spot Price FOB (Dollars per Barrel)'] /42
df_dados['Europe Brent Spot Price FOB (Dollars per Gallon)'] = df_dados['Europe Brent Spot Price FOB (Dollars per Barrel)'] /42
```

```python
df_dados.tail()
```

```python
# Cálculo do Spread
```

```python
lista_cols  = df_dados.columns
for i in range(len(lista_cols)):
    if (i>1) and (i<13):
      novo_nome = "Spread_"+lista_cols[i]
      print(i, novo_nome)
      df_dados[novo_nome] = df_dados[lista_cols[i]] / df_dados[lista_cols[11]]
```
